```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(15, 6))
numbers = np.arange(10)
numbers
```

```python
plt.plot(numbers)
plt.show()
```

```python
x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)
```

```python
plt.figure(figsize=(15, 6))
plt.plot(x, y, linewidth=2, linestyle='--', color='y', marker='o')
plt.title('Sin(X) function')
plt.xlabel('X (radians)')
plt.ylabel('sin(X)')
plt.grid(color = 'k', linestyle='--')
plt.show()
```

```python
x = np.arange(0, 4 * np.pi, 0.1)
sin = np.sin(x)
cos = np.cos(x)

plt.figure(figsize=(15, 6))

plt.plot(x, sin,
         linewidth=1.5, 
         linestyle='-.',
         color='#8E44AD',
         alpha=0.7,
         marker='h', 
         label='sin(x)')

plt.plot(x, cos,
         linewidth=1.5, 
         linestyle=':',
         color='c',
         alpha=0.7,
         marker='v', 
         label='cos(x)')

plt.title('Sin and Cosine Functions', fontsize=20, fontweight='bold', color='black')
plt.xlabel('X (radians)', fontsize=14)
plt.ylabel('Function Value', fontsize=14)

plt.legend(loc='lower left', fontsize='x-large', title='Functions')

plt.grid(color='tomato', alpha=0.5, linestyle='--', linewidth=1)

plt.show()
```
